# 01 · Setup & Authentication

This notebook walks through getting access to the **[Bigdata.com](https://bigdata.com)**
API and making your first authenticated request.

## 1. Register
Look for your invitation email from **invitations@bigdata.com** and follow the
link to create your account. (If you don't have an invite, contact your
Bigdata.com representative or support@bigdata.com.)

## 2. Log in to the app
Sign in at **[app.bigdata.com](https://app.bigdata.com)**.

## 3. Open the Developer Platform
Go to the **[Developer Platform](https://platform.bigdata.com)** — your hub for
API keys, interactive playgrounds, and usage/spend monitoring.

## 4. Create an API key
In the Developer Platform open **API Keys** and create a new key. Copy it
immediately — you will not be able to see the full value again.

## 5. Store the key as an environment variable

Keep keys **out of code and notebooks**. Set an environment variable instead.

```bash
# macOS / Linux (add to ~/.zshrc or ~/.bashrc to persist)
export BIGDATA_API_KEY="your_api_key_here"
```
```powershell
# Windows PowerShell
setx BIGDATA_API_KEY "your_api_key_here"
```
Restart your terminal / Jupyter kernel after setting it.

````python
import os
import requests

# Your API key is read from the BIGDATA_API_KEY environment variable.
# Never hard-code keys in a notebook you plan to share or commit.
API_KEY = os.environ["BIGDATA_API_KEY"]

# Bigdata exposes two hosts:
#   - api.bigdata.com    -> Knowledge Graph & Search Service (deterministic APIs)
#   - agents.bigdata.com -> Research Agent & Workflows (AI agents, streamed)
API_BASE = "https://api.bigdata.com"
AGENTS_BASE = "https://agents.bigdata.com"

# The key must be sent in the X-API-KEY header on every request.
HEADERS = {"X-API-KEY": API_KEY, "Content-Type": "application/json"}
````

## 6. Authentication model

Every request must include your key in the **`X-API-KEY`** header. Equivalent cURL:

```bash
curl --request POST \
  --url https://api.bigdata.com/v1/search \
  --header 'Content-Type: application/json' \
  --header 'X-API-KEY: <api-key>' \
  --data '{"query": {"text": "Microsoft earnings call highlights"}}'
```

## 7. Verify your key works
We call a lightweight endpoint — the Knowledge Graph sectors list — to confirm the key is accepted.

````python
resp = requests.get(f"{API_BASE}/v1/knowledge-graph/companies/sectors", headers=HEADERS, timeout=30)

if resp.status_code == 200:
    sectors = resp.json()["results"]
    print("\u2705 Authentication successful. Available sectors:")
    for s in sectors:
        print("  -", s)
elif resp.status_code in (401, 403):
    print("\u274c Key rejected. Check that BIGDATA_API_KEY is set correctly.")
else:
    print("Unexpected status:", resp.status_code, resp.text[:300])
````
